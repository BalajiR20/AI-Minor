# Importing libraries

Data visualisation:
* Data:Iris dataset
* Tools: Pandas, Numpy, Matplotlib, Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.plotting import andrews_curves
from pandas.plotting import parallel_coordinates
```

```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
data
```

# Loading Data

```python
df = pd.DataFrame(data=data.data, columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.target.value_counts()
```

```python
df.columns.isna().sum()
```

# Data Visualisation

```python
sns.scatterplot(df,x='sepal length (cm)',y='sepal width (cm)',hue='target',palette='viridis')
```

```python
sns.scatterplot(df,x='petal length (cm)',y='petal width (cm)',hue='target',palette="pastel")
```

```python
sns.pairplot(df,hue='target',palette='viridis')
```

```python
sns.jointplot(df,x='petal length (cm)',y='petal width (cm)',hue='target')
```

```python
sns.FacetGrid(df, hue="target").map(plt.scatter, "sepal length (cm)", "sepal width (cm)").add_legend()
plt.show()
```

```python
sns.boxplot(df,x="target",y="petal length (cm)")
```

```python
ax= sns.boxplot(x="target", y="petal length (cm)", data=df,palette='viridis')
ax= sns.stripplot(x="target", y="petal length (cm)", data=df, jitter=True, edgecolor="gray",color='r')
plt.show()
```

```python
sns.violinplot(x="target", y="petal width (cm)", data=df,palette="Paired")
```

```python
sns.violinplot(x="target", y="sepal length (cm)",  palette={"violet","pink","teal"}, data=df)
plt.show()
```

```python
sns.FacetGrid(df, hue="target").map(sns.kdeplot, "petal length (cm)").add_legend()
plt.show()
```

```python
sns.pairplot(df,hue='target',palette='viridis',diag_kind="hist")
```

```python
sns.jointplot(df,x='sepal length (cm)',y='sepal width (cm)',ratio=10, kind='hex',color='green')
```

```python
andrews_curves(df, "target", colormap="rainbow")
plt.show()
```

```python
parallel_coordinates(df, "target", colormap="cool")
plt.show()
```

```python
sns.countplot(x=df["target"],palette='Paired')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(df,x='petal length (cm)',bins=50,kde=True,color='y')
```

```python
sns.histplot(df,x='petal width (cm)',bins=50,kde=True,color='teal')
```
